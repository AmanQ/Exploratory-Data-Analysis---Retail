```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

KDE represents the data using a continuous probability density curve in one or more dimensions.

```python

plt.figure(figsize=(9,7))
sns.kdeplot(dataset['Sales'],color='green',label='Sales',shade=True)
sns.kdeplot(dataset['Profit'],color='red',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.legend()
```

Sales in RHS of graph overtakes the profit

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));
sns.countplot(dataset['Category'],ax=axs[0][0])
sns.countplot(dataset['Segment'],ax=axs[0][1])
sns.countplot(dataset['Ship Mode'],ax=axs[1][0])
sns.countplot(dataset['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=20)
axs[0][1].set_title('Segment',fontsize=20)
axs[1][0].set_title('Ship Mode',fontsize=20)
axs[1][1].set_title('Region',fontsize=20)

plt.tight_layout()
```

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (10,10))
sns.distplot(dataset['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(dataset['Profit'], color = 'green',  ax = axs[0][1])
sns.distplot(dataset['Quantity'], color = 'orange',  ax = axs[1][0])
sns.distplot(dataset['Discount'], color = 'blue',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```

Statewise analysis of Profit Discount and sell

```python

dataset['State'].value_counts().head(10)
```

```python
df_state= dataset.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

Vermont: Highest Profit
, Ohio: Lowest Profit

```python
df_state['Sales'].plot(kind='pie',
                        figsize = (20,20),
                        autopct='%1.1f%%',
                        startangle=90,     # start angle 90 (Africa)
                        shadow=True)
plt.title('State wise analysis of Sale',fontsize=20)
```

Highest amount of sales= Wyoming(11.8%) and
Lowest amount of sales= South Dakota(0.8%)

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5))
plt.title('State wise analysis of Discount', fontsize=20)
```

**Citywise Analysis of the Profit**

```python
df_city= dataset.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head()
```

```python
#1.Low Profit
df_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'red')
plt.title('City wise analysis of 30 cities with Lowest profit')
```
